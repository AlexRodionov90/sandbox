```python
import matplotlib.pyplot as plt


labels = ["A", "B", "C", "D"]
sizes = [40, 25, 20, 15]
colors = ["#ff9999", "#66b3ff", "#99ff99", "#ffcc99"]
fig, ax = plt.subplots()


plt.pie(sizes, labels=labels, wedgeprops=dict(width=0.7), autopct="%1.1f%%", startangle=90, colors=colors)
plt.title("Распределение")
legend_obj = plt.legend()
legend_obj.set_draggable(True)


plt.savefig('1.png', transparent=True)
```
